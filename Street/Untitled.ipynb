### import csv

```python
import pandas as pd

post = pd.read_csv('LOLA_adj_postlemma.csv')
pre = pd.read_csv('LOLA_adj_prelemma.csv')

lemma = post
lemma = lemma.append(pre, sort=True)
```

```python
adjectif = 0 
for i in range(54037):
    if lemma['adjectif_pre'].values[i] == "vert": 
        adjectif+=lemma['total'].values[i]
adjectif
```

```python
n = lemma.groupby('adjectif_pre').sum().sort_values(['total'], ascending=False)
```

```python
n.to_csv('adjectiftotal.csv', sep='\t', encoding='utf-8')
```
